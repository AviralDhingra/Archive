```python
a = np.array([[1,2,3] , [5,6,7]])
b = np.array([[3,4,5] , [7,7,7]])
c = a*b
print(c)

# * refers to scalar multiplication
```

```python
a = np.array([[1,2,3] , [5,6,7]])
b = np.array([[3,4] , [7,7] , [8,8]])
c = np.matmul(a,b)
print(c)

# the col of the first matrix should match the row of the second matrix
```

```python
a = np.array([[1,2,3] , [5,6,7]])
b = np.array([[3,4] , [7,7] , [8,8]])
c = np.dot(a,b)
print(c)
```
