```python
def gender_value_check():
    gender_values = df['gender']
    for i in range(len(gender_values)):
        # print(f'_{gender_values[i]}_')
        if gender_values[i] != 'M' or gender_values[i] != 'F':
            print(f'Row {i}: {gender_values[i]}')
```

```python
l = ['Test Value']
for i in range(len(df.index)):
    current = df.loc[i, 'ssc_b']
    for j in range(len(l)):
        if current == l[j]:
            break
        elif current != l[j] and j == len(l):
            l.append(current)
    # print()
    # print(type(df.loc[i, 'ssc_b']))
```
